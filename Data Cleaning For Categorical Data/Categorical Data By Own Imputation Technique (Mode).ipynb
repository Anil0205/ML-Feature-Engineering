# With Sklearn

```python
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('house_price.csv')
```

```python
df.head()
```

```python
c_a=df.select_dtypes(include='object')
```

```python
c_a.head()
```

```python
reg = SimpleImputer(strategy = 'constant' ,fill_value = 'Missing')
```

```python
#reg.fit(c_a)
```

```python
reg.fit(c_a['FireplaceQu'].values.reshape(-1,1))
```

```python
c_a['FireplaceQu']=reg.transform(c_a['FireplaceQu'].values.reshape(-1,1))
```

```python
c_a['FireplaceQu'].isnull().sum()
```
